```python
red = np.array([200,50,50])
blue = np.array([50,50,200])
green = np.array([127,127,0])
yellow = np.array([210,210,0])
white = np.array([210,210,210])
black = np.array([-50,-50,-50])

colors = np.stack((red,yellow,green,blue,white,black))
color_dict = {0:'R',1:'Y',2:'G',3:'B',4:'W',5:'K'}

def identify_color(rgb_val):
    c_list = []
    for i in colors:
        c_list.append(np.sum((i-rgb_val)**2))
    
    return color_dict[c_list.index(min(c_list))]
```

```python
def get_centers(im, im_bin):
    centers = {}
    num_labels, labels = cv2.connectedComponents(im_bin)
    for i in range(num_labels):
        x = np.int(np.mean(np.where(labels==i)[0]))
        y = np.int(np.mean(np.where(labels==i)[1]))
        c = identify_color(im[x,y])
        if c != 'K':
            centers[x,y] = c
    return centers
```

```python
A = np.array([])
for i in range(4):
    k = str(i)
    x0,y0 = x['p'+k]
    x_t0,y_t0 = x_t['p'+k]
    p = np.array([[-x0, -y0, -1, 0, 0, 0, x0*x_t0, y0*x_t0, x_t0],
               [0, 0, 0, -x0, -y0, -1, x0*y_t0, y0*y_t0, y_t0]])
    if i == 0:
        A = p
    else:
        A = np.concatenate((A,p),axis=0)

print(A)
```

```python
U,S,Vt = np.linalg.svd(A)
print('Size of A matrix: ',A.shape)
print('Size of U matrix: ',U.shape)
print('Size of S matrix: ',S.shape)
print('Size of Vt matrix: ',Vt.shape)
```

```python
V = Vt.T
H = V[:,-1].reshape(3,3)
print('Size of H matrix: ',H.shape)
```
